### 1.Import the necessary library. Load the dataset into the pandas data frame. 

```python
import seaborn as sbn
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
import numpy as np
df = pd.read_csv("https://raw.githubusercontent.com/TrainingByPackt/Data-Science-with-Python/master/Chapter01/Data/german_credit_data.csv")
```

### 2.Plot the boxplot of the age column. To do so, use boxplot() function from the seaborn library

```python
sbn.boxplot(df['Age'])
```

### 3.Find the Inter quartile range of Age column from 25th quartile and 75th quartile.

```python
Q1 = df["Age"].quantile(0.25)
Q3 = df["Age"].quantile(0.75)
IQR = Q3 - Q1
print(IQR)
```

### 4.Find the Upper Fence and Lower Fence 

```python
Lower_Fence = Q1 - (1.5 * IQR)
Upper_Fence = Q3 + (1.5 * IQR)
```

```python
print(Lower_Fence)
print(Upper_Fence)
```

### 5.Print all the data below the Lower fence and above the Upper fence

```python
df[(df["Age"] < Lower_Fence) |(df["Age"] > Upper_Fence)]
```

### 6.Filter out the outlier data and print only the potential data. To do so, just negate the above result using ~ operator

```python
df[~((df["Age"] < Lower_Fence) |(df["Age"] > Upper_Fence))]
```
